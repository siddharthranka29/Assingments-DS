```python
import pandas as pd
import numpy as np
import warnings
warnings.simplefilter('ignore')
```

```python
data=pd.read_csv("D:\\anime.csv")
```

```python
data
```

```python
#Data Preprocessing & Feature Extraction
data.describe()
```

```python
data.isnull().sum()
```

```python
data["genre"].mode()
```

```python
data["genre"]=data["genre"].fillna(data["genre"].mode()[0])
```

```python
data["type"]=data["type"].fillna(data["type"].mode()[0])
data["rating"]=data["rating"].fillna(0)
data.isnull().sum()
```

```python
data1=pd.pivot_table(data=data,index='anime_id',columns='genre',values='rating')
data1.fillna(0,inplace=True)
data1
```

```python
#Recommendation System & Evaluation
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df_sim=cosine_similarity(data1)
df_sim
```

```python
sim_df=pd.DataFrame(df_sim)
sim_df
```

```python
sim_df.index=data.anime_id.unique()
sim_df.columns=data.anime_id.unique()
sim_df
```

```python
np.fill_diagonal(df_sim,0)
sim_df.idxmax()[0:10]
```
